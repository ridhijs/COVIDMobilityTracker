```python
import random
gmr.reset_index(drop=True)
gmr['aveMob'] = 0
gmr['mockCon'] = 0
columns = []
for col in gmr:
    columns.append(col)
for row in gmr.iterrows():
    a = gmr.loc[row[0],columns[5]]
    b = gmr.loc[row[0],columns[6]]
    c = gmr.loc[row[0],columns[7]]
    d = gmr.loc[row[0],columns[8]]
    e = gmr.loc[row[0],columns[9]]
    f = gmr.loc[row[0],columns[10]]
    
    gmr['aveMob'][row[0]] = round(sum([a,b,c,d,e,(f*-1)]), 2)
```

```python
mockConDict = dict()
for county in gmr['sub_region_2'].unique():
    mockConDict[county] = random.randint(0,70)

for row in gmr.iterrows():
    gmr['mockCon'][row[0]] = mockConDict[gmr['sub_region_2'][row[0]]]
```

```python
combine = covid.merge(gmr, left_on='county',right_on='sub_region_2', how='inner')
combine
```

```python
combine['cases'].dropna(inplace=True)
```

```python
combine['cases']
```

```python
def CIRI(active,a,n):
    Rzero = 3.28 #average R0 estimate by Liu et al, 2020 - https://academic.oup.com/jtm/article/27/2/taaa021/5735319
    k = 4 #k is how many connections does the place have.
    #I am assuming a constant value because of the lack of this information, for now.
#     normCIRI = (x-min(x))/(max(x)-min(x))
    try:
        goelSharma = (Rzero * ((1 + (1+a*k)*n)/(1+n))) 
    except:
        goelSharma = 0
    #Based on equation by Goel and Sharma, 2020 - https://arxiv.org/pdf/2004.13015.pdf
    normGS = goelSharma / 1761.36     #We are diving CIRI by the highest possible value to get an index from 0:1
    
    CIRI = active * normGS/active
    
    return CIRI

#CIRI(70, 45)
```

```python
combine['active']
```

```python
combine['aveCIRI'] = 0.0
combine['retailCIRI'] = 0.0
combine['groceriesCIRI'] = 0.0
combine['parksCIRI'] = 0.0
combine['transitCIRI'] = 0.0
combine['workplacesCIRI'] = 0.0
combine['residentialCIRI'] = 0.0
CIRIvalues = dict()

for row in combine.itertuples():
    cases = row.cases
    a = row.mockCon
    n = row.aveMob
    retail = row.retail_and_recreation_percent_change_from_baseline
    groceries = row.grocery_and_pharmacy_percent_change_from_baseline
    parks = row.parks_percent_change_from_baseline
    transit = row.transit_stations_percent_change_from_baseline
    workplaces = row.workplaces_percent_change_from_baseline
    residencial = row.residential_percent_change_from_baseline
    
    CIRIvalues[row.Index] = [CIRI(cases,a,n),
                            CIRI(cases, a,retail),
                            CIRI(cases,a,groceries),
                            CIRI(cases,a,parks),
                            CIRI(cases,a,transit),
                            CIRI(cases,a,workplaces),
                            (1 - CIRI(cases,a,residencial))]
    
    
#combine.to_csv('NYwithCIRI.csv')
```

```python
for row in combine.iterrows():
    CIRIvalue = CIRIvalues[row[0]]
    
    combine['aveCIRI'][row[0]] = CIRIvalue[0]
    combine['retailCIRI'][row[0]] = CIRIvalue[1]
    combine['groceriesCIRI'][row[0]] = CIRIvalue[2]
    combine['parksCIRI'][row[0]] = CIRIvalue[3]
    combine['transitCIRI'][row[0]] = CIRIvalue[4]
    combine['workplacesCIRI'][row[0]] = CIRIvalue[5]
    combine['residentialCIRI'][row[0]] = 1 - CIRIvalue[6]
```

```python
ciriDF = pd.DataFrame.from_dict(CIRIvalues, orient = 'index', columns = ['aveCIRI', 'retailCIRI', 'groceriesCIRI', 'parksCIRI', 'transitCIRI','workplacesCIRI','residentialCIRI'])
```

```python
combine['aveCIRI'] = 0.0
combine['retailCIRI'] = 0.0
combine['groceriesCIRI'] = 0.0
combine['parksCIRI'] = 0.0
combine['transitCIRI'] = 0.0
combine['workplacesCIRI'] = 0.0
combine['residentialCIRI'] = 0.0
```

```python
for row in ciriDF.itertuples():
    combine['aveCIRI'][row[0]] = row.aveCIRI
    combine['retailCIRI'][row[0]] = row.retailCIRI
    combine['groceriesCIRI'][row[0]] = row.groceriesCIRI
    combine['parksCIRI'][row[0]] = row.parksCIRI
    combine['transitCIRI'][row[0]] = row.transitCIRI
    combine['workplacesCIRI'][row[0]] = row.workplacesCIRI
    combine['residentialCIRI'][row[0]] = row.residentialCIRI
```
